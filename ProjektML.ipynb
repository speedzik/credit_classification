```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from imblearn.under_sampling import RandomUnderSampler
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
```

Trenowanie i preprocessing modelu.

W nastÄ™pnym kroku budujemy strumienie, ktÃ³re zakodujÄ… nasze cechy nieliczbowe, przeskalujÄ… cechy liczbowe oraz dodatkowo zmnniejszÄ… wymiar caÅ‚ego datasetu.

```python





numerical = X.select_dtypes(include = 'number').columns
categorical = X.select_dtypes(exclude = 'number').columns

categorical_transformer = Pipeline(steps=[("encode", OneHotEncoder(handle_unknown = 'ignore')),('pca', PCA())])
numerical_transformer = Pipeline(
   steps=[("imputation", SimpleImputer()), ("scaling", MinMaxScaler()), ('pca', PCA())]
)

preprocessor = ColumnTransformer(
   transformers=[
     ("num", numerical_transformer, numerical),
     ("cat", categorical_transformer, categorical)
   ]
)

```

Klasyfikatory, ktÃ³re uÅ¼yÅ‚em.

```python
classifiers = {
    'RandomForest': RandomForestClassifier(),
    'SVM': SVC(),
    'LogisticRegression': LogisticRegression(max_iter=1000, multi_class='multinomial')
}
```

Parametry dla RandomSearch'a

```python
param_grids = {
    'RandomForest': {
        'preprocessor__num__pca__n_components': [10, 15, 20],
        'classifier__n_estimators': [50, 100],
        'classifier__max_depth': [10, 20, None]
    },
    'SVM': {
        'preprocessor__num__pca__n_components': [10, 15, 20],
        'classifier__C': [0.1, 1, 10],
        'classifier__kernel': ['linear', 'rbf'],
        'decision_function_shape': ['ovr', 'ovo']
    },
    'LogisticRegression': {
        'preprocessor__num__pca__n_components': [10, 15, 20],
        'classifier__C': [0.1, 1, 10],
        'classifier__solver': ['liblinear', 'saga']
    }
}
```

```python
best_estimators = {}
```

```python
for name, classifier in classifiers.items():
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', classifier)
    ])
    
    random_search = RandomizedSearchCV(pipeline, param_distributions=param_grids[name], n_iter=50, cv=5, n_jobs=-1, scoring='accuracy', random_state=42)

    random_search.fit(X_train_resampled, y_train_resampled)
    
    best_estimators[name] = random_search.best_estimator_
    print(f"Best parameters for {name}: {random_search.best_params_}")
    print(f"Best cross-validation score for {name}: {random_search.best_score_}")

for name, estimator in best_estimators.items():
    print(f"Evaluating {name} on the test set")
    y_pred = estimator.predict(X_test)
    print(classification_report(y_test, y_pred))
```
